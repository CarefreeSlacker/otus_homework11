```
from keras.models import Sequential
from keras.layers import Dense, Activation

import sklearn.datasets as ds
import matplotlib.pyplot as plt
import pylab as pl
import numpy as np

```

```
# Для начала сгенерируем датасет на 1000 элементов 

(data, labels) = ds.make_moons(n_samples=1000, noise=0.05)

pl.figure(figsize=(10, 5))

pl.scatter(data[:, 0], data[:, 1], c=labels, cmap=pl.cm.PuOr, lw=0, s=30)
plt.xlabel('X')
plt.ylabel('Y')

pl.show()
```

```
def model_constructor(output_class_amount):
    return Sequential([
        Dense(8, input_shape=(2,), activation='relu'),
        Dense(8, activation='relu'),
        Dense(output_class_amount, activation='softmax')
    ])

model = model_constructor(1)


# For a binary classification problem
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```
model.fit(data, labels, epochs=50, verbose=1)
# По какой-то неясной причине. Если на выходе функция активации softmax, как указанно в задании, то точность всегда 0.5, что соответствует случайному совпадению.
# Попробуем с другой функцией.
```

```
model = model_constructor(1)

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```
model.fit(data, labels, epochs=60, verbose=1)
# С функцией активации Sigmoid на выходе удаётся норально обучить модель.
```

```
# Визуализируем полученные результаты

def plot_model_results(decision, X, Y):
    x_min = X[:, 0].min()
    x_max = X[:, 0].max()
    y_min = X[:, 1].min()
    y_max = X[:, 1].max()

    x, y = np.mgrid[x_min:x_max:.01, y_min:y_max:.01]
    pos = np.empty(x.shape + (2,))
    pos[:, :, 0] = x
    pos[:, :, 1] = y
    pos_lin = pos.reshape(pos.shape[0] * pos.shape[1], 2)
    pred_lin = decision(pos_lin)
    pred = pred_lin.reshape(pos.shape[0], pos.shape[1])

    fig, ax = plt.subplots()
    ax.set_xlim((x_min, x_max))
    ax.set_ylim((y_min, y_max))
    ax.contourf(x, y, pred, cmap='Reds', alpha=0.5)
    ax.contour(x, y, pred, levels=[0.5])

    colors = ['yellow' if y_i else 'blue' for y_i in Y]
    plt.scatter(X[:, 0], X[:, 1], c=colors)
    plt.show()
```

```
# Отлично. Модель обучена.
plot_model_results(model.predict, data, labels)
```

```
import mnist
```

```
from keras.layers import Flatten

elements_limit = 10000
train_values = mnist.train_images()[:elements_limit]
train_labels = mnist.train_labels()[:elements_limit]
test_values = mnist.train_images()[elements_limit:(elements_limit + 1000)]
test_labels = mnist.train_labels()[elements_limit:(elements_limit + 1000)]
```

```
def mnist_model_constructor(output_classes_count):
    return Sequential([
        Dense(10, input_shape=(28, 28), activation='relu'),
        Dense(10, activation='relu'),
        Flatten(),
        Dense(output_classes_count, activation='softmax')
    ])

mnist_model = mnist_model_constructor(10)

mnist_model.compile(optimizer='adam',
                    loss='sparse_categorical_crossentropy',
                    metrics=['accuracy'])

mnist_model.fit(train_values, train_labels, epochs=40, verbose=1)
# Модель обучена
```

```
# Провеврим точность обученной модели
import sklearn.metrics as smt

def calculate_accuracy(test, pred):
    accuracy = smt.adjusted_rand_score(test, pred)
    print('accuracy {0}'.format(accuracy))
    return
```

```
pred_labels = mnist_model.predict_classes(test_values)

calculate_accuracy(test_labels, pred_labels)
# Точность всего ~80%, возможно модель переобучена.
# Попробуем обучить заново и проверить после обучения снова, при меньшем значении acc при обучении.
```

```
mnist_model = mnist_model_constructor(10)

mnist_model.compile(optimizer='adam',
                    loss='sparse_categorical_crossentropy',
                    metrics=['accuracy'])

mnist_model.fit(train_values, train_labels, epochs=80, verbose=1)
```

```
pred_labels = mnist_model.predict_classes(test_values)

calculate_accuracy(test_labels, pred_labels)
# Сюда это не вошло, прогонял несколько раз при меньшем acc, точность была ниже. 
# Решил прогнать на большем количестве эпох. Как видно - точность не сильно увеличилась. Попробуем обучить дата сет на большем количестве данных.
```

```
# Весь дата сет состоит из 60000 записей. Возьмём из них 50000, протестируем классификацию на 5000

elements_limit = 50000
test_limit = 5000

train_values = mnist.train_images()[:elements_limit]
train_labels = mnist.train_labels()[:elements_limit]
test_values = mnist.train_images()[elements_limit:(elements_limit + test_limit)]
test_labels = mnist.train_labels()[elements_limit:(elements_limit + test_limit)]
```

```
mnist_model = mnist_model_constructor(10)

mnist_model.compile(optimizer='adam',
                    loss='sparse_categorical_crossentropy',
                    metrics=['accuracy'])

mnist_model.fit(train_values, train_labels, epochs=30, verbose=1)
```

```
pred_labels = mnist_model.predict_classes(test_values)

calculate_accuracy(test_labels, pred_labels)
# Точность удалось увеличить до 90% Результат отличный. Задание можно считать выполненным.
```

```
# Попробуем докрутить модель. Улучшим оптимизатор и увеличить количество эпох.
import keras

mnist_model = mnist_model_constructor(10)

mnist_model.compile(optimizer=keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False),
                    loss='sparse_categorical_crossentropy',
                    metrics=['accuracy'])

mnist_model.fit(train_values, train_labels, epochs=60, verbose=1)
```

```
pred_labels = mnist_model.predict_classes(test_values)

calculate_accuracy(test_labels, pred_labels)
# Это было долго. Но результат аналогичный. Тем не менее задание считаю выполненным.
```
